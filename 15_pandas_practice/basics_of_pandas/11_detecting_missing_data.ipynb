Detecting Missing Data

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', None, 'David'],
    'Age': [25, None, 22, 23],
    'City': ['Delhi', 'Mumbai', 'Bangalore', None]
}

df = pd.DataFrame(data)

#if there is any null value or NaN

print(df.isnull())
print(df.isna())

#if rows are not null or NaN

print(df.notnull())
print(df.notna())

#sum of all the null values

print(df.isnull().sum())
print(df.isna().sum())

#sum of all the not null values

print(df.notnull().sum())
print(df.notna().sum())

#all null values
print(df.isna().all())

# if there is any null value/(atleast 1)
print(df.isna().any())

# return rows with atleast one missing value
df[df.isna().any(axis=1)]
```

### Practice Tasks

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Amit', 'Priya', 'John', np.nan, 'Neha', 'Ravi', np.nan],
    'Age': [25, np.nan, 22, 28, 30, np.nan, 27],
    'City': ['Delhi', 'Mumbai', np.nan, 'Chennai', np.nan, 'Kolkata', 'Delhi']
}

df = pd.DataFrame(data)
print(df)

#q1 - how many missing values are in each column?

print(df.isnull().sum(axis=0))

#q2 - How many total missing values are there in the whole DataFrame?

print(df.isna().sum().sum())

#q3 : Which rows have at least one missing value?

print(df[df.isna().any(axis=1)])

#q4 : Which columns have no missing values at all?

print(df.notna().all(axis=0))

#q5 : Print all rows that are completely filled (no missing values).

print(df[df.notna().all(axis=1)])

#q6 : Print only those rows where all values are missing.

print(df[df.isna().all(axis=1)])

#q7 : Print the column names where at least one value is missing.

print(df.columns[df.isna().any(axis=0)])
```
