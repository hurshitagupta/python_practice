### PATIENT VITALS MONITOR

Goal: You have to :

- **Dummy Dataset** : Create a 3D array of shape (10,7,3) where 10 is no of patients, days are 7 and vitals are 3

- **Calculate Per-Patient Averages** : to calculate average temperature, pulse and oxygen for each patient

- **Detect Fever Days** : Find which days they had temp > 37.5°C

- **Oxygen Drop Alert** : Find which patients ever had O2 < 92%


```python
import numpy as np

#task1 : creating 3D array of size (10,7,3)
#temp_range: 36.0-39.5
temp_arr = 3.5 * np.random.random_sample(size=70).reshape(70,1) + 36.0
# print(temp_arr)

#pulse_range : 60-110
pulse_arr = np.random.randint(60,111,size=70).reshape(70,1)
# print(pulse_arr)

#oxygen-range : 88-100
oxygen_arr = np.random.randint(88,101,size=70).reshape(70,1)
# print(oxygen_arr)

patients_data = np.hstack([temp_arr,pulse_arr,oxygen_arr])
p_data = patients_data.reshape(10,7,3)
print("Data for 10patients, 7days and 3vitals:\n",p_data)


# task2 = calculate avg temp, pulse and oxygen for each patient
#1. avg temp
temps = p_data[:,:,0]
temp_mean = np.mean(temps,axis=1).reshape(10,1)
# print(temp_mean)

#2. avg pulse
pulse = p_data[:,:,1]
pulse_mean = np.mean(pulse,axis=1).reshape(10,1)
# print(pulse_mean)

#3. avg oxygen
oxygen = p_data[:,:,2]
oxygen_mean = np.mean(oxygen, axis=1).reshape(10,1)
# print(oxygen_mean)

averages = np.hstack([temp_mean,pulse_mean,oxygen_mean])
print("Average temperature, pulse and oxygen for each patient:\n",averages)


#task3 : Find which days they had temp > 37.5°C
# print(p_data[:,:,0] > 37.5)
fever = np.where(p_data[:,:,0] > 37.5)
days = fever[1]+1
print("Days when patients had fever:",days)

#task4 : Find which patients ever had O2 < 92%
oxygen_drop = p_data[:,:,2]
drop = oxygen_drop < 92
patient = np.where(np.any(drop,axis=1)) 
patient_num = patient[0] + 1
print("Patient numbers who fall ill are:",patient_num)
```
